```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv('911.csv')
df.info() #twp township
```

```python
#top 5 count of zip column
df['zip'].value_counts().head(5)
```

```python
#unique count of title
df['title'].nunique() #count of unique 
```

```python
df['reason']=df['title'].apply(lambda x: x.split(':')[0]) #get EMS from EMS:test
df['reason'].value_counts().head(2) #two most reason to call 911
```

```python
sns.countplot(x='reason',data=df,palette='viridis')
```

```python
#convert timestamp string to datetime obj
df['timeStamp']=pd.to_datetime(df['timeStamp'])
df['timeStamp'].iloc[0].hour
```

```python
df['timeStamp'].iloc[0]
```

```python
df['hour']=df['timeStamp'].apply(lambda x: x.hour)
df['month']=df['timeStamp'].apply(lambda x: x.month)
df['day_of_week']=df['timeStamp'].apply(lambda x: x.dayofweek)
df.head()
```

```python
dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
df['day_of_week']=df['day_of_week'].map(dmap) #map index to string as given in dmap
df['day_of_week']
```

```python
sns.countplot(x='day_of_week',data=df,hue='reason')
plt.legend(bbox_to_anchor=(1.05, 1),loc=2,borderaxespad=0.) #drawing legend outside of graph
```

```python
sns.countplot(x='month',data=df,hue='reason',palette='viridis')

# To relocate the legend It is missing some months! 9,10, and 11 are not there.
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
bymonth=df.groupby('month').count()
bymonth
```

```python
bymonth['lat'].plot() #drop from 8 to 12 but we can fill the gap
```

```python
sns.lmplot(x='month',y='twp',data=bymonth.reset_index()) #linear model plot, reset_index() makes month as index
#no of call goes down from 0 to 12 month , shaded area are error it grows as month grows
#trend going down and up, crashes at month 8, 
```

```python
#get date from timestamp 
df['date']=df['timeStamp'].apply(lambda x: x.date())
df.head()
```

```python
df.groupby('date').count()['lat'].plot()
plt.tight_layout() #date text overlap removal
```

```python
df[df['reason']=='Traffic'].groupby('date').count()['lat'].plot() #traffic reason plot of date , and lat count
plt.tight_layout() #date text overlap removal
```

```python
#group by with unstack method
#df.groupby(['day_of_week','hour']).count()#create multi level index
#df.groupby(['day_of_week','hour']).count()['reason']#create multi level index, unstack this and make this into matrix form
dayofhour=df.groupby(['day_of_week','hour']).count()['reason'].unstack()# unstack this and make this into matrix form
dayofhour
```

```python
plt.figure(figsize=(12,6)) #make figure heatmap bigger
sns.heatmap(dayofhour,cmap='coolwarm')
```

```python
sns.clustermap(dayofhour,cmap='coolwarm')
```

```python
from pandas_datareader import data, wb
df = pd.read_pickle('all_banks.pkl')
df
```

```python
from pandas_datareader import data, wb
import datetime
```

```python
start=datetime.datetime(2006,1,1) #year,month,date
end=datetime.datetime(2016,1,1)
BAC = df.BAC
```

```python
bac=BAC[(BAC.index>=start) & (BAC.index <= end)]
c=df.C[(df.C.index>=start) & (df.C.index<=end)]
gs=df.GS[(df.GS.index>=start) & (df.GS.index<=end)]
jpm=df.C[(df.JPM.index>=start) & (df.JPM.index<=end)]
ms=df.C[(df.MS.index>=start) & (df.MS.index<=end)]
wfc=df.C[(df.WFC.index>=start) & (df.WFC.index<=end)]
```

```python
tickers = ['bac', 'c', 'gs', 'jpm', 'ms', 'wfc']
```

```python
bac
```

```python
bank_stocks=pd.concat([bac,c,gs,jpm,ms,wfc],axis=1,keys=tickers) #concat along columns becuase all shared same index
#keys argument are tickers
bank_stocks.head()
```

```python
bank_stocks.columns.names = ['Bank Ticker','Stock Info']
bank_stocks.head() #first index label , first row called bank ticker, 2nd row call stock info
```

```python
bank_stocks.index
```

```python
#cross section xs, select data at a particular level of a MultiIndex.
for tick in tickers: 
    print(tick,'\t',bank_stocks[tick]['Close'].max())
```

```python
bank_stocks.xs(key='Close',axis=1,level='Stock Info').max() #column axis, same logic return as above, inner level called stock info
 
```

```python
returns=pd.DataFrame() #empty data frame
for tick in tickers:
    returns[tick+' Return'] = bank_stocks[tick]['Close'].pct_change() #pct_change do P(t)/P(t-1) -1
returns
#first is null becuase nothing is pass behind it
```

```python
import seaborn as sns
sns.pairplot(returns[1:]) #0th is null that why start from row 1 to last
#c : city group is stright becuase it is crashed
```

```python
#what dates each bank stock had the best and worst single day returns
returns.min() #return each bank min value from all dates 
```

```python
returns['bac Return'].idxmin() #return the date of minimum value, index of min value
```

```python
returns.idxmin()
```

```python
returns.idxmax()
```

```python
# a basic measurement of risk need riskiness of a stock is the standard deviation of the returns of the stock.
#if you have a very large standard deviation of the returns it means your stock is going up and down a lot 
returns.std() #citiy group have highest deviation so risky stock
```

```python
#for 2015
returns['2015-01-01':'2015-12-31'].std() #grabe slice of index, do standar deviation, well forgoes lowest risk
#risky: but Morgan Stanley or Bac
```

```python
sns.distplot(returns['2015-01-01':'2015-12-31']['ms Return'],color='green',bins=100) #Morgan Stanley distplot
#it deviate .06 from average (distribution plot)
```

```python
sns.distplot(returns['2008-01-01':'2008-12-31']['c Return'],color='red',bins=100)
#quite a bit more stretched out especially here you're almost that .6 not .06 so city group deviate 10 times more
```
