```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer
cancer=load_breast_cancer()
cancer.keys()
```

```python
cancer['DESCR']
```

```python
df_feat = pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
df_feat.info()
```

```python
df_feat
```

```python
df_target = pd.DataFrame(cancer['target'],columns=['Cancer'])
```

```python
df_target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_feat, np.ravel(df_target), test_size=0.30, random_state=101)
```

```python
from sklearn.svm import SVC #Train the Support Vector Classifier
model = SVC()
model.fit(X_train,y_train)

```

```python
predictions = model.predict(X_test)
from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(y_test,predictions))
```

```python
print(classification_report(y_test,predictions))
```

```python
#t may also help to actually normalize the data as well.We can search for parameters using a GridSearch!
#do is search for the best parameters using a grid search
#Now a grid search allows you to find the right parameters such as like what C or gamma values to use and finding those right parameters is usually a tricky task.
# the idea of creating a grid of parameters is trying out all the best possible combinations it's called a grid search.
from sklearn.model_selection import GridSearchCV #takes in a dictionary that describes the parameters that should be tried in a model to train the grid
```

```python
#C controls the cost of misclassification on the training data.
#A large C value gives you low bias and high variance low bias because you penalize the cost of misclassification for larger c value
#with a smaller C value. You're not going to penalize that cost as much. So it gives you a higher bias and a lower variance
#gamma parameter has to do with the free parameter of the Gaussian radio radial basis function 
#kernel='rbf' radio basis function , a small Gamla means a Gaussian for large variance.
# a large Gamla value is going to lead to a high bias and low variance in the model or vice versa.
#So if the Gamla is large then variance is small implying that support vector does not have a widespread influence.
#Well go ahead and call C as one of the parameters you want to test.
#And I'm going to test out quite a range I'll say from 0.1 1 10 100 and then 1000 and then let's go ahead
param_grid = {'C': [0.1,1, 10, 100, 1000], 'gamma': [1,0.1,0.01,0.001,0.0001], 'kernel': ['rbf']} 
grid = GridSearchCV(SVC(),param_grid,refit=True,verbose=3) 
#The higher the number the more verbose and verbose just means the text output of the description of the process.
```

```python
grid
```

```python
# May take awhile!
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_estimator_
```

```python
grid_predictions = grid.predict(X_test)
```

```python
print(confusion_matrix(y_test,grid_predictions))
```

```python
print(classification_report(y_test,grid_predictions))
```
