```python
#K Means Clustering will allow us to cluster and label data in an unsupervised machine learning algorithm.is an unsupervised learning algorithm that tries to cluster data based on their similarity. 
#Unsupervised learning means that there is no outcome to be predicted, and the algorithm just tries to find patterns in the data
#going to to attempt to group similar clusters together in your data. 
# k value as we just saw to perform k means clustering. We have to decide how many clusters we expect in the data.
# elbow method. :First of all you compute the sum of squared error otherwise known as s s e. For some values of k for example 2 4 6 8
#The sum of squared errors is defined as the sum of the square distance between each member of the cluster and its centroid.
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8,random_state=101)
#This basically generates blob like groups of data. std: standard deviation, 
```

```python
data #tuple data[0] numpy array,  basically your
#number of samples and then two columns of features so if we actually grab the shape of this it's 200 by two.
data[0].shape #within this this random data it will actually have four blobs in it.
```

```python
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow') #then grab all the rows in the first column of data[0] and plot those against data
#zero all the rows in the second column. c: center it is in data[1] first row
```

```python
data[1]
```

```python
#this algorithm going to just start off by randomly assigning each observation to a cluster and then finds a centroid of each cluster.
#Then the algorithm will iterate through the two steps. Remember it reassigned data points to the cluster whose centroid is closest.
#It calculates the new central for each cluster and then it just repeats that over and over again until the cluster variation can't be reduced any further.
kmeans = KMeans(n_clusters=4)
kmeans.fit(data[0])# fit to the features and the features are just data zero where we plotted earlier.
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(10,6)) #subplot of 1 row 2 columns, 2 asix
ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='rainbow')
ax2.set_title("Original")
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```
