```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('USA_Housing.csv')
df.head()
```

```python
df.describe()
```

```python
df.columns
```

```python
x=df[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
       'Avg. Area Number of Bedrooms', 'Area Population']] #features
```

```python
y=df['Price'] #target variables
```

```python
#We want to split our data into a training set for the model in a testing set in order to test the model
#once it's been trained.
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4, random_state=101)
#using touple and packing, test size. It's just a percentage of your data set that you want to be allocated to the test size.
#specific set of random splits on your data because this train split is going to occur randomly in order to make sure that your random split is the same as my random split

```

```python
#train the model 
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(x_train,y_train) #train or fit my model or my training data and fit.
```

```python
lm.intercept_ #Let's evaluate the model by checking out it's coefficients and how we can interpret them.
#the coefficients are going to relate to each feature in our dataset.
```

```python
lm.coef_ #these coefficients relates to the columns in x or x train.
```

```python
coeff_df = pd.DataFrame(lm.coef_,x.columns,columns=['Coefficient']) #x.columns as index, 
coeff_df
#if we hold all other features fixed a one unit increase in average area income,  
#is associated with an increase of $21 and 52 cents in price.
#one year increase in the average area ,price increase by a hundred sixty four thousand eight hundred eighty three dollars 
```

```python
preditions=lm.predict(x_test)
preditions #These are the predicted prices of the house. y_test are correct price of the house
```

```python
y_test
```

```python
#how far off are the predictions from the tests prices the actual prices. using scatter plot
plt.scatter(y_test,preditions) #compare y_test with prediction
```

```python
#residuals remember are the difference between the actual values why test and the predicted values
sns.distplot((y_test-preditions)) #have normally distributed residuals.
#If this is not normally distributed,check to see if a linear regression model was the correct choice for the data set 
```

```python
#three common evaluation metrics for regression problems.
#Mean Absolute Error (MAE) is the mean of the absolute value of the errors:
#Mean Squared Error (MSE) is the mean of the squared errors:,MSE is more popular than MAE, because MSE "punishes" larger errors, which tends to be useful in the real world.
#Root Mean Squared Error (RMSE) is the square root of the mean of the squared errors:, is even more popular than MSE, because RMSE is interpretable in the "y" units.
#y means target unit, minimazie this we want to minimize all these errors in order to create the best model.
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, preditions))
print('MSE:', metrics.mean_squared_error(y_test, preditions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, preditions)))
```
