```python
import seaborn as sns
tip=sns.load_dataset('tips') #save as data frame
tip.head()
```

```python
# show the distribution of a univariate (one variable) set of observations 
sns.displot(tip['total_bill'],bins=30)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tip) #combine 2 different distribution plot, x and y data are compared on tip dataset 
#total bill on x higher bill more tips
```

```python
sns.jointplot(x='total_bill',y='tip',data=tip,kind='hex') #hexagonal represantion of above
```

```python
sns.pairplot(tip,hue='sex',palette='coolwarm') #join plot of every possible numerical column, categorical column: group so sex is column here
#pallete is color
```

```python
sns.rugplot(tip['total_bill']) #raws a dash mark for every points on this uniform or unique variant distribution essentially one single variable.
#more dash means more points , at 40-50 less dash means in histogram you will see less height
```

```python
#KDE plots stand for kernel density estimation plots 
sns.displot(tip['total_bill'],kde=True)
```

```python
import numpy as np
np.random.randn(25)
```

```python
sns.barplot(x='sex',y='total_bill',data=tip) #categorical data plot, shows avg total bill
```

```python
sns.barplot(x='sex',y='total_bill',data=tip, estimator=np.std) # estimator is group by function , here we can say do by standar deviation func

```

```python
sns.countplot(x='sex',data=tip) #count the occurences
```

```python
#to show that distribution of categorical data a box plot is also sometimes known as a box and whisker plot.
sns.boxplot(x='day',y='total_bill',data=tip,hue='smoker')
#This box plot shows the core tiles of the data set while the whiskers extend to show the rest of the 
#distribution except for these points that are outside of these whiskers.
#box plot split by smoker status 25% of first
```

```python
sns.violinplot(x='day',y='total_bill',data=tip,hue='smoker') #plot all the components that correspond
#to actual data points and it's essentially showing the kernel density estimation of the underlying distribution
```

```python
sns.stripplot(x='day',y='total_bill',data=tip, jitter=True,hue='sex') #jitter=true, add a little bit of random noise to separate some of these stacked points.

#split will give more clear picture on two category by sex
```

```python
#combine idea of stripplot and voilen plot
#they don't actually scale that well to very large numbers both in terms of the ability to show all the points.
#And it also takes a lot of computation to begin to arrange them.


sns.swarmplot(x='day',y='total_bill',data=tip)
```

```python
fp=flights.pivot_table(index='month',columns='year',values='passengers')
fp
```

```python
sns.heatmap(fp,cmap='coolwarm',linecolor='black',linewidths=1) #magma from dark to light
```

```python
#cluster version of heatmap
sns.clustermap(fp,cmap='coolwarm',standard_scale=1) #similarity within data, see months are not in order
#try to show columns and rows that are similar to each other so I can see now that 1959 and 1960 are very similar to each other.
#so aug and july
#if you wanted to normalize this you can called standard scale=1,normalize this to 0 to 1
#So it looks like this November January and February is kind of early winter months are much closer to low passengers 
```
