```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#You just know that you need to use these features that are unknown to you as far as what they actually
#produce target class ,because the KNN(K nearest neighbours) classifier predicts the class
#of a given test observation by identifying the observations that are nearest to it.
#of a given test observation by identifying the observations that are nearest to it. represent in order to predict a target class 1 or 0
df = pd.read_csv("Classified Data.csv",index_col=0)
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df.drop('TARGET CLASS',axis=1))
scaled_features = scaler.transform(df.drop('TARGET CLASS',axis=1)) #transformation or scaled version of data, return array of data
scaled_features
```

```python
#do is use that scaled features variable to recreate a feature.
df_feat = pd.DataFrame(scaled_features,columns=df.columns[:-1]) #remove last column name
df_feat.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(scaled_features,df['TARGET CLASS'],
                                                    test_size=0.30)
```

```python
#Remember that we are trying to come up with a model to predict whether someone will TARGET CLASS or not. We'll start with k=1.
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(x_train,y_train)
```

```python
pred = knn.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,pred))
```

```python
print(confusion_matrix(y_test,pred))
```

```python
#Let's go ahead and use the elbow method to pick a good K Value:
error_rate = []

# Will take some time
for i in range(1,40):
    
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train,y_train)
    pred_i = knn.predict(x_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

```python
#Here we can see that that after arouns K>23 the error rate just tends 
#to hover around 0.06-0.05 Let's retrain the model with that and check the classification report!

# NOW WITH K=28
knn = KNeighborsClassifier(n_neighbors=28)

knn.fit(x_train,y_train)
pred = knn.predict(x_test)

print('WITH K=28')
print('\n')
print(confusion_matrix(y_test,pred))
print('\n')
print(classification_report(y_test,pred))
```
