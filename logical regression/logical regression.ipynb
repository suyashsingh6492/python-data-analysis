```python
#fill the missing age: called imputation take the average age and fill in the missing values with the average age.
#check the average age passenger class.
plt.figure(figsize=(10,7)) #larger plot
sns.boxplot(x='Pclass',y='Age',data=train) #1 premium class, 3 cheap class see 1class paseenger are older 
```

```python
def impute_age(cols):
    Age=cols[0]
    Pclass=cols[1]

    if(pd.isnull(Age)):
        if Pclass==1 :
            return 37
        elif Pclass==2 :
            return 29
        else:
            return 24

    return Age


        
```

```python
train['Age']= train[['Age','Pclass']].apply(impute_age,axis=1) #list > apply on column
#train.drop('Cabin',axis=1,inplace=True) #drop Cabin info on column , apply on train data
train.dropna(inplace=True) #drop all missing value 
train.head()
```

```python
#convert categorical features into what is known as a dummy variable,Otherwise our machine learning algorithm won't be able to directly take in those features as inputs
#sex column here we have a categorical feature of male or female convert this to 0 and 1
#in order to encode that information in a way that a machine learning algorithm can understand it. And this is known as creating a dummy variable 
#convert embark column State or City S/C
sex = pd.get_dummies(train['Sex'],drop_first=True) #convert categorical variables into dummy or otherwise known as indicator variables.
#if our machine learning algorithm gets fed both columns the machine learning algorithm will
#immediately know that hey if it's zero female I can predict perfectly that it's going to be one male and this is going to be an issue known as multi-collinearity 
#a bunch of column will be perfect predictors of a another column in order to avoid this. so drop first column using drop_first=True
embark = pd.get_dummies(train['Embarked'],drop_first=True) #Q and S will tell the exact value
sex
```

```python
train.drop(['Sex','Embarked','Name','Ticket'],axis=1,inplace=True) #these column are no longer use
train = pd.concat([train,sex,embark],axis=1) #concate data by column
train
```

```python
train.drop('PassengerId',axis=1,inplace=True)
```

```python
train
```

```python
#y is the column that you're trying to predict the actual label.
y=train['Survived']
x=train.drop('Survived', axis=1) #all column except survived
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=101) #test size 30%

```

```python
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()
logmodel.fit(x_train,y_train)
```

```python
predictions = logmodel.predict(x_test)
```

```python
#We can check precision,recall,f1-score using classification report!
#read it yourself often confusion matrix.
from sklearn.metrics import classification_report 
print(classification_report(y_test,predictions)) #83% accurecy
```
