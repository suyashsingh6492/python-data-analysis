```python
df=df.drop('zipcode',axis=1)
df=df.drop('date',axis=1)
df.head()
```

```python
X=df.drop('price',axis=1)
y=df['price']
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size = 0.30,random_state =101)
```

```python
#do scaling to fit the test
#Remember, we wanna do the scaling post-split. That way we only fit to the training set to prevent data leakage from the test set.
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
#scaling actually has this convenience feature of transforming and then fitting at one step
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test) #dont fit we don't want to assume prior information about our test set.
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X_train.shape #So it looks like we have 19 incoming features
#and that's probably a good range to then have 19 neurons
```

```python
model=Sequential()
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(1))
```

```python
model.compile(optimizer='adam',loss='mse') #mean square error
#since this is a regression problem, we're choosing a continuous label, such as price.
```

```python
# TensorFlow may complain if you don't pass in a numeric array because it can't work with pandas series or DataFrames that well.
#we wanna be checking against our test set
#because this is a larger dataset, we're gonna feed in our data in batches. And we're gonna call batch_size of 128.
model.fit(x=X_train,y=y_train,validation_data=(X_test,y_test),batch_size=128,epochs=400) # it will only use the training data as it's updating the weights and biases
#It's very typical to do batch_sizes in powers of two.
#So 64, 128, 256. The smaller the batch_size, the longer training is going to take          
```

```python
losses=pd.DataFrame(model.history.history)
losses
```

```python
losses.plot() #then there is no increase so far in the validation set. later
#If you saw this orange line begin to spike after some
#of these epochs, so it started to go up and up and up
#that means you're overfitting to the training data because now you have a much larger loss on your validation data.
```

```python
#let's do some evaluation on our test data.
from sklearn.metrics import mean_squared_error,mean_absolute_error,explained_variance_score
predictions=model.predict(X_test)
```

```python
mean_squared_error(y_test,predictions) #large value
```

```python
np.sqrt(mean_squared_error(y_test,predictions))

```

```python
mean_absolute_error(y_test,predictions)
```

```python
#Early stopping is something we can do with Keras to automatically stop training based on a loss condition
#on the validation data  Dropout layers are layers that can be added to turn off neurons during training to prevent overfitting.
# each dropout layer will drop, or turn off, a user-defined percentage of neuron units
# in the previous layer every batch. So that means certain neurons don't have their weights or biases affected during a batch,
import sklearn.datasets
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
breast_cancer_dataset
```

```python
X=df.drop('label',axis=1).values
y=df['label'].values
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size = 0.25,random_state =101)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
X_train.shape
```

```python
model=Sequential() 
model.add(Dense(30, activation='relu'))# add 30 neurons
model.add(Dense(15, activation='relu'))
#because this is a binary classification problem. , we want the last activation to be sigmoid
# one neuron, outputting something between a zero and one
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam')
```

```python
model.fit(x=X_train,y=y_train,epochs=600,validation_data=(X_test,y_test),verbose=0)
#epoch is hight
```

```python
model.history.history
```

```python
losses=pd.DataFrame(model.history.history)
losses
```

```python
losses.plot() #and this is a perfect example of overfitting.,
#during the first couple of epochs,
#both validation and training loss are both decreasing. That's good. That means we haven't yet overfit to our training data
#and as we go along, we're decreasing loss in both the validation set and the training set.
# at a certain point in time, notice that our training loss here in blue, is still going down.
#our validation loss is beginning to increase.
#That basically tells us that we're overfitting to our training dataset.
```

```python
# use tensorflow keras callbacks to actually, based on your validation loss, stop the training
model=Sequential() 
model.add(Dense(30, activation='relu'))# add 30 neurons
model.add(Dense(15, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam')
```

```python
from tensorflow.keras.callbacks import EarlyStopping
EarlyStopping
```

```python
early_stop=EarlyStopping(monitor='val_loss',mode='min',verbose=1,patience=25) #mode: what are you actually trying to do here?
#minimaize, or maximize accuracy is something you want to maximize.
#If our metric is loss, essentially the reverse of accuracy, loss is something you want to minimize.
#That means we'll wait 25 epochs even after we've kind of detected a stopping point because of noise that could occur.
model.fit(x=X_train,y=y_train,epochs=600,validation_data=(X_test,y_test),callbacks=[early_stop])
#epoch is hight#it stopped training after 75
```

```python
model_log=pd.DataFrame(model.history.history)
model_log.plot()
```

```python
#So dropout layers will essentially turn off a percentage of neurons randomly.
# use tensorflow keras callbacks to actually, based on your validation loss, stop the training
from tensorflow.keras.layers import Dropout
model=Sequential() 
model.add(Dense(30, activation='relu'))# add 30 neurons
model.add(Dropout(.5))#rate is the probability that you're going to randomly turn off the actual neurons.
#1 means 100% neuron,
#In our case, because we haven't specified a batch size, half
#of those neurons in this layer of 30 are going to be turned off so their weights and biases won't be getting updated.
model.add(Dense(15, activation='relu'))
model.add(Dropout(.5))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam')
```

```python
early_stop=EarlyStopping(monitor='val_loss',mode='min',verbose=1,patience=25) #mode: what are you actually trying to do here?
model.fit(x=X_train,y=y_train,epochs=600,validation_data=(X_test,y_test),callbacks=[early_stop])
```

```python
model_log=pd.DataFrame(model.history.history)
model_log.plot() #better behavior, 
```
