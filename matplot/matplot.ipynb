```python
import matplotlib.pyplot as plt
%matplotlib inline 
#this will allow to see plot inside the jupiter notebook
```

```python
import numpy as np
x=np.linspace(0,5,11) #liner space 0 to 5 grab 11 points
```

```python
x
y=x**2
y
```

```python
#functional method
plt.plot(x,y,'r--') #3rd argument is color red and line style --, 
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
plt.show()
```

```python
plt.subplot(1,2,1) #no of rows, no of column, 3rd is plot number
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
plt.show()
plt.close()
```

```python
#object oriented method
fig=plt.figure()#this object is imaginary blank canvas, 
axes=fig.add_axes([.1,.1,.8,.8]) #takes left,bottom, width , height range from 0 to 1 bacisally percentages of blank canvas
# .1 is left of axes, bottom is .1, width =.8, height=.8
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')

fig
plt.close()
```

```python
fig=plt.figure()
axes1=fig.add_axes([.1,.1,.8,.8])
axes2=fig.add_axes([.2,.5,.4,.3]) # start from 20 percent of left, 50 percent of bottom, take 40% of canvas of width, 30% canvas size in height
axes1.plot(x,y)
axes2.plot(y,x)
axes2.set_title('small plot')
fig
plt.close()
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2) #2 column 1 row, based on arg, auto calling add axes
plt.tight_layout() #fixing overlap 
axes
#as we have 4 graph, we have 4 axes, just print axes before to see how many graph available 
for current_ax_obj in axes:
    for cur_axes in current_ax_obj:
        cur_axes.plot(x,y)
plt.tight_layout() #fixing overlap 
fig
```

```python
fig.clear()
plt.close()
```

```python
fig=plt.figure(figsize=(5,4),dpi=100) #figsize : figure size width, height in inches,dpi:dot per inch,pixel per inch
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X Squared') #what this plot means is label
ax.plot(x,x**3,label='X Cubed')
#ax.legend(loc=10) #generate lagend by label , localtion is where legend should be, 0:best, 1:right
ax.legend(loc=(1.5,.5)) #x,y coordinate to specify the legend
fig
```

```python
fig.savefig('my_graph.png',dpi=200) #save figure in png format
fig.clear()
plt.close()
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='red',linewidth=3,alpha=.5,linestyle='-',
        marker='o',markersize=20,markerfacecolor='yellow',markeredgewidth=3,markeredgecolor='green') 
#alpha arg: how transparent line is, linestyle : .,-,--,steps
#marker used when we have few data points
ax.set_xlim([0,2]) #want to show x from 0 to 1
ax.set_ylim([0,12])
fig
```
